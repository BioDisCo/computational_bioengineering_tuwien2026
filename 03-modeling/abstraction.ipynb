# Abstraction

In the following sections we summarize abstractions from the very basics of quantum mechanics to chemical reaction networks.

Choosing the right abstraction for a problem is an essential skill in computational bioengineering: too coarse ones oversee critical phenomena to explain a phenomenon, too fine ones make it untracktable or obfuscate the essentials.
The right abstraction is not easy to find - and one may change abstractions many times during the process of finding a solution.

We hope that this chapter helps in gaining a basic understanding of typical abstractions.
